## Lambda

Forma general:

**lambda** argumentos : expresión

La expresión es lo que se devuelve

```python
suma = lambda a, b: a + b
print(suma(3, 4))
```

equivalente a:

```python
def suma(a, b):
    return a + b

print(suma(8, 3))
```

Mas de un parametro

```python
operacion = lambda a, b: a if a > b else b
print(operacion(4, 7))
```

Lambda con if else

```python
es_par = lambda x: True if x % 2 == 0 else False
print(es_par(6))
print(es_par(5))
```

Lambda + listas

```python
nums = [1, 2, 3, 4]
doble = list(map(lambda x: x * 2, nums))
print(doble)
```

Lambda + filter

```python
nums = [1, 2, 3, 4, 5, 6]
pares = list(filter(lambda x: x % 2 == 0, nums))  # filter conserva solo los True
print(pares)
```

Lambda + sort

```python
palabras = ["sol", "mariposa", "luz", "estrella"]
palabras.sort(key=lambda x: len(x))  # len ordena por longitud
print(palabras)
```

Lambda + diccionario

```python
operaciones = {
    "suma": lambda a, b: a + b,
    "resta": lambda a, b: a - b
}

print(operaciones["suma"](5, 3))
print(operaciones["resta"](5, 3))
```

```python
'''dada una lista de elementos, cree una función que devuelva un diccionario sin los elementos duplicados.'''

def quitar_duplicados(lista):
    resultado = {}
    for elemento in lista:
        if elemento not in resultado:
            resultado[elemento] = elemento
    return resultado

datos = ["a", "b", "a", "c", "b", "d"]
print(quitar_duplicados(datos))
```

```python
'''dada una lista de elementos, cree una función que devuelva una tupla sin los elementos duplicados.'''

def quitar_duplicados(lista):
    resultado = []
    for elemento in lista:
        if elemento not in resultado:
            resultado.append(elemento)
    return tuple(resultado)

datos = ["a", "b", "a", "c", "b", "d"]
print(quitar_duplicados(datos))
```

```python
'''dada un diccionario, cree una función que devuelva una lista sin los elementos duplicados.'''

def valores_unicos(diccionario):
    resultado = []
    for valor in diccionario.values():
        if valor not in resultado:
            resultado.append(valor)
    return resultado

datos = {"a": 1,"b": 2,"c": 1,"d": 3,"e": 2}
print(valores_unicos(datos))
```

```python
# con set

def valores_unicos(diccionario):
    return list(set(diccionario.values()))

datos = {"a": 1,"b": 2,"c": 1,"d": 3,"e": 2}
print(valores_unicos(datos))
```
